```python
import numpy as np

#part 1a
N = 8
n = np.arange(N)
x = ((0.5)**n)*np.cos(np.pi*n/3)
```

```python
#part 1b
delta = np.zeros(N)
delta[0] = 1
```

```python
#part 1c
z = np.arange(1, N+1)
X = (0.5/(1-0.5*(z*np.exp(1j*np.pi/3))**(-1))) + (0.5/(1-0.5*(z*np.exp(-1j*np.pi/3))**(-1)))
```

```python
import matplotlib.pyplot as plt
from scipy import signal

y = signal.lfilter(np.array([1]), X, delta)

plt.figure()
plt.subplot(211)
plt.stem(x)
plt.subplot(212)
plt.stem(y)
```

```python
#part 2a
import matplotlib.pyplot as plt

a = [1, -0.9]
b = [1, 0]
plt.figure()
z,p,_ = signal.tf2zpk(b, a)
plt.scatter(np.real(p), np.imag(p), c='b', marker='x')
plt.scatter(np.real(z), np.imag(z), c='r', marker='o')
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5])
plt.title("Pole-Zero Plot")
```

```python
#part 2b
N = 8
delta = np.zeros(N)
delta[0] = 1

a = [1, -0.9]
b = [1, 0]

fig, axs = plt.subplots(2,1)
w, h = signal.freqz(b, a)
axs[0].plot(w, 20 * np.log10(abs(h)))
axs[0].set_title('Magnitude response')
axs[0].set_xlabel('Normalized frequency [radians / pi]')
axs[0].set_ylabel('Amplitude [dB]')

axs[1].plot(w, np.angle(h))
axs[1].set_title('Phase response')
axs[1].set_xlabel('Normalized frequency [radians / pi]')
axs[1].set_ylabel('Phase [radians]')
plt.show()
```
